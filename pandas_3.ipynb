```python
d1 = {
    'key1':[1,2,3,4,5,6],
    'key2':[7,8,9,10,11,"Manish"],
    'key3':[13,14,15,16,18,19],
    'key4':[20,21,22,23,24,25]
}
```

```python
import pandas as pd
```

```python
pd.DataFrame(d1, index = ['a','b','c','d','e','f'])
```

```python
df1 = pd.DataFrame(d1)
```

```python
d2 = {
    'key1':[1,"FSDS",3,4,5,6],
    'key2':[7,8,9,10,11,"Manish"],
    'key3':[13,14,"Bisht",16,18,19],
    'key4':["ineuron",21,22,23,24,25]
}
```

```python
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how = 'inner', on = 'key1')
```

```python
pd.merge(df1, df2, how = 'left', on = 'key4')
```

```python
pd.merge(df1, df2, how = 'right', on = 'key4')
```

```python
pd.merge(df1, df2, how = 'outer', on = 'key4')
```

## join operations


```python
d10 = {
    'key11':[1,2,3,4,5,6],
    'key22':[7,8,9,10,11,"Manish"],
    'key33':[13,14,15,16,18,19],
    'key44':[20,21,22,23,24,25]
}
```

```python
d20 = {
    'key1':[1,2,3,4,5,6],
    'key2':[7,8,9,10,11,"Manish"],
    'key3':[13,14,"Bisht",16,18,19],
    'key4':["ineuron",21,22,23,24,25]
}
```

```python
df10 = pd.DataFrame(d10, index = [1,2,3,4,5,6])
```

```python
df20 = pd.DataFrame(d20, index = [1,2,3,4,5,'mani'])
```

```python
df10
```

```python
df20
```

```python
df10.join(df20, how = 'right')
```

```python
df10.join(df20, how = 'inner')
```

```python
df10.join(df20, how = 'outer')
```

```python
df10.join(df20)
```

- Concat operation is done on columns (Horizontally or Vertically)....
-  join operation is done on index...
- merge operation can be understood as similar to venn diagram merger (inner, outer, left, right)
