```python
import cufflinks as cf
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.randn(100,3), columns = ["A","B","C"])
```

```python
cf.go_offline()
```

```python
df.iplot()
```

```python
df.iplot(x="A", y="B", mode="markers", size= 6)
```

```python
df
```

```python
df.iplot(kind='bubble', x="A", y = "B", size= "C")
```

```python
tip = pd.read_csv('tips.csv')
```

```python
tip
```

```python
tip[['total_bill','tip']].iplot(x='total_bill',y='tip', mode="markers", size= 6, xTitle= "total_bill", yTitle="tips")
```

```python
tip.iplot(kind='scatter3d', x = 'total_bill', y = 'tip', z = 'size')
```

```python
tip.iplot(kind='bubble3d', x = 'total_bill', y = 'tip', z = 'size', size='tip')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = tip)
```

```python
sns.relplot(x='total_bill', y = 'tip', data = tip, hue='smoker')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = tip, style='size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = tip, style='size', hue='smoker')
```
